```python
import pandas as pd

# Step 1: Load the dataset
ferdousi_path = '/content/drive/MyDrive/Deep Learning/DL_HW4/Q3 Materials/ferdousi.txt'
with open(ferdousi_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Step 2: Ignore the first two lines
lines = lines[2:]

# Step 3: Split the poem into verses
verses = []
current_verse = ""
for line in lines:
    if line.strip() == "":
        if current_verse:
            verses.append(current_verse)
            current_verse = ""
    else:
        current_verse += line.strip() + " "

# Step 4: Create input-output pairs
input_verses = verses[:-1]
output_verses = verses[1:]

# Step 5: Split the dataset into training and testing sets
split_ratio = 0.8  # 80% for training, 20% for testing
split_index = int(len(input_verses) * split_ratio)

train_input = input_verses[:split_index]
train_output = output_verses[:split_index]

test_input = input_verses[split_index:]
test_output = output_verses[split_index:]

# Print to see the results
print("Head of the whole dataset:")
print(list(zip(input_verses, output_verses))[:5])

print("\nHead of the training dataset:")
print(list(zip(train_input, train_output))[:5])

print("\nHead of the testing dataset:")
print(list(zip(test_input, test_output))[:5])
```
